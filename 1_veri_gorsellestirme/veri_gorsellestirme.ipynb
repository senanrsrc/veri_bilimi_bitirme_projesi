
# 1. Veri Görselleştirme Projesi (50_Startups.csv)

- Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar.
- R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.
- Profit (Kâr): Şirketin elde ettiği toplam kâr.
- Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.
- Profit (Kâr): Şirketin elde ettiği toplam kâr.
- State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).


#### → Kütüphaneler içeri aktarıldı.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Görselleştirme ayarları
plt.style.use('default')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 10
```

#### → 50_startups.csv dosyası DataFrame'e yüklendi.
Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.

```python
data = pd.read_csv('50_Startups.csv')
print(data.head())
```

## Görev 1 : R&D Harcaması ile Kar Arasındaki İlişkiyi Scatter Plot ile Gösterme

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['R&D Spend'], data['Profit'], alpha=0.7, color='blue', s=50)
plt.xlabel('R&D Harcaması ($)', fontsize=12)
plt.ylabel('Kâr ($)', fontsize=12)
plt.title('R&D Harcaması vs Kâr İlişkisi', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Korelasyon ( ilişki) hesaplama
correlation_rd = data['R&D Spend'].corr(data['Profit'])
print(f"R&D Harcaması ve Kâr arasındaki korelasyon: {correlation_rd:.4f}")
```

## Görev 2 : Yönetim Harcaması ile Kar Arasındaki İlişkiyi Scatter Plot İle Gösterme

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Administration'], data['Profit'], alpha=0.7, color='green', s=50)
plt.xlabel('Yönetim Harcaması ($)', fontsize=12)
plt.ylabel('Kâr ($)', fontsize=12)
plt.title('Yönetim Harcaması vs Kâr İlişkisi', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Korelasyon hesapla
correlation_admin = data['Administration'].corr(data['Profit'])
print(f"Yönetim Harcaması ve Kâr arasındaki korelasyon: {correlation_admin:.4f}")
```

## Görev 3 :  Eyaletlere Göre Ortalama Karları Bar Chart ile Karşılaştırma

```python
ortalama = data.groupby('State')['Profit'].agg(['mean']).round(2)
print("Eyaletlere göre ortalama kar")
print(ortalama)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(ortalama.index, ortalama['mean'],
               color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Eyalet', fontsize=12)
plt.ylabel('Ortalama Kâr ($)', fontsize=12)
plt.title('Eyaletlere Göre Ortalama Kâr', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')
for i, ort in enumerate(ortalama['mean']):
    plt.text(i, ort, f'{ort:,.0f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

## Görev 4 : R&D, Yönetim ve Pazarlama Harcamalarının Dağılımını Boxplot ile Karşılaştırma

```python
harcama = pd.DataFrame({
    'R&D Harcaması': data['R&D Spend'],
    'Yönetim Harcaması': data['Administration'],
    'Pazarlama Harcaması': data['Marketing Spend'] })
plt.figure(figsize=(12, 6))
harcama.boxplot()
plt.title('Harcama Türlerinin Dağılımını Karşılaştırma', fontsize=14, fontweight='bold')
plt.ylabel('Harcama Miktarı ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
